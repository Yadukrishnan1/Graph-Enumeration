```python
import lammps_logfile
import matplotlib.pyplot as plt

log = lammps_logfile.File("read_reax_385k.txt")

step = log.get("Step")
temp = log.get("Temp")
dens = log.get("Density")

plt.figure(figsize=(15,4))
plt.subplot(131)
plt.plot(step, temp,'.')
plt.ylabel('T (K)')
plt.xlabel('Time (fs)')
plt.subplot(132)
plt.plot(step, dens,'.')
plt.ylabel('Density (g/cc)')
plt.xlabel('Time (fs)')
plt.subplot(133)
plt.plot(temp[:301], dens[:301],'k-')
plt.plot(temp[301:], dens[301:],'r-')
plt.ylabel('Density (g/cc)')
plt.ylabel('T (K)')
plt.show()
```

```python
vol=log.get("Volume")
pot=log.get("PotEng")
plt.figure(figsize=(15,4))

plt.subplot(131)
plt.plot(step, pot,'.')
plt.subplot(132)
plt.plot(step, vol,'.')
plt.subplot(133)
```

```python
import lammps_logfile
import matplotlib.pyplot as plt

log385 = lammps_logfile.File("read_reax_385k.txt")
log425 = lammps_logfile.File("read_reax_425k.txt")

step425 = log425.get("Step")
temp425 = log425.get("f_TempAve")

step385 = log385.get("Step")
temp385 = log385.get("f_TempAve")


plt.figure(figsize=(6,4))

plt.plot(step385, temp385,'g.-', label='385 K')
plt.plot(step425, temp425,'r.-', label='425 K')
plt.ylabel('T (K)')
plt.xlabel('Time (fs)')
plt.ylim(300,500)
plt.legend()
plt.savefig('T_behav_ring.png')
plt.show()
```
